```python
%matplotlib inline

import bandits as bd
```

```python
n_trials = 500
n_experiments = 100
```

# 研究Thompson Sampling的性质

```python
n_arms = 10  # 选择越多，前期纯探索的代价越大

bandit = bd.BernoulliBandit(n_arms, t=3*n_trials)

agents = [
    bd.BetaAgent(bandit, bd.GreedyPolicy())
]
env = bd.Environment(bandit, agents, label='Bayesian Bandits')
scores, optimal = env.run(n_trials, n_experiments)
```

```python
env.plot_results(scores, optimal)
```

# 研究Thompson Sampling和Frequentist实验的区别

## 二选一，经典实验

```python
n_arms = 2  
bandit = bd.BernoulliBandit(n_arms, t=3*n_trials)

agents = [
    bd.agent.FrequentistAgent(bandit, bd.FrequentistPolicy(n_trials/2)),  # 一半探索，一半利用
    bd.BetaAgent(bandit, bd.GreedyPolicy())
]
env = bd.Environment(bandit, agents, label='Bayesian Bandits Vs Frequentist Experiment (2 Choice)')
scores, optimal = env.run(n_trials, n_experiments)
```

```python
env.plot_results(scores, optimal)
```

## 多维选择

```python
n_arms = 10  
bandit = bd.BernoulliBandit(n_arms, t=3*n_trials)
env = bd.Environment(bandit, agents, label='Bayesian Bandits Vs Frequentist Experiment (10 Choice)')
scores, optimal = env.run(n_trials, n_experiments)
```

```python
env.plot_results(scores, optimal)
```
