```python
%matplotlib inline

import bandits as bd
```

```python
n_arms = 10  # 选择越多，前期纯探索的代价越大
n_trials = 1000
n_experiments = 100
bandit = bd.BernoulliBandit(n_arms, t=3*n_trials)
```

```python
agents = [
    bd.agent.FrequentistAgent(bandit, bd.FrequentistPolicy(500)),
    bd.BetaAgent(bandit, bd.GreedyPolicy())
]
env = bd.Environment(bandit, agents, label='Bayesian Bandits')
scores, optimal = env.run(n_trials, n_experiments)
```

```python
# 前500个step，平均优选率1/n；后500个step，接近但未能打到100%，因为有一定概率没有选中最优解
env.plot_results(scores, optimal)
```
